```python
def lotto(chasu):
    url="http://www.dhlottery.co.kr/common.do?method=getLottoNumber&drwNo="+str(chasu)
    
    result_data = urlopen(url)
    result = result_data.read()
    
    data = json.loads(result)
    data_1 = pd.DataFrame.from_dict(data, orient='index')
    data_1 = data_1.transpose() ## 행과 열을 바꾼거 
    
    return data_1
```

```python
df_e_cols=["totSellamnt","returnValue","drwNoDate","firstWinamnt","firstPrzwnerCo","firstAccumamnt","drwNo","drwtNo1","drwtNo2","drwtNo3","drwtNo4","drwtNo5","drwtNo6","bnusNo"]
df_k_cols=["누적금액","실행결과","추첨일자","1등당첨금","1등당첨인원","1등누적당첨금","회차","당첨번호1","당첨번호2","당첨번호3","당첨번호4","당첨번호5","당첨번호6","보너스번호"]
rows = []
```

```python
for name in df_e_cols:
    print("'" + name + "'", ": row.loc[0, '", name, "'],")
```

```python
for number in tqdm(range(1071,1090)):
    row=lotto(number)

    if row.loc[0,"returnValue"]=="fail":
        pass
    else:
        rows.append({'누적금액' : row.loc[0,'totSellamnt'],
        '실행결과' : row.loc[0,'returnValue'],
        '추첨일자' : row.loc[0,'drwNoDate'],
        '1등당첨금' : row.loc[0,'firstWinamnt'],
        '1등당첨인원' : row.loc[0,'firstPrzwnerCo'],
        '1등누적당첨금' : row.loc[0,'firstAccumamnt'],
        '회차' : row.loc[0,'drwNo'],
        '당첨번호1' : row.loc[0,'drwtNo1'],
        '당첨번호2' : row.loc[0,'drwtNo2'],
        '당첨번호3' : row.loc[0,'drwtNo3'],
        '당첨번호4' : row.loc[0,'drwtNo4'],
        '당첨번호5' : row.loc[0,'drwtNo5'],
        '당첨번호6' : row.loc[0,'drwtNo6'],
        '보너스번호' : row.loc[0,'bnusNo']})

lotto = pd.DataFrame(rows,columns=df_k_cols)     
lotto
```
